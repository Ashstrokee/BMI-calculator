```python
name  = input("Enter Your Name: ")
weight = int(input("Enter Your Weight in kg: "))
height = int(input("Enter Your height in centimeters: "))
BMI = (weight) / (height * height/10000)
if BMI>0:
    if BMI<18.5:
        print(name +", C'mon,you're underweight")
    elif BMI<=24.9:
        print(name +",You're spot on ")
    elif BMI<=29.9:
        print(name +",You're overweight,hit the gym") 
    elif BMI<=34.9:
        print(name +",Obesity Stage 1 ") 
    elif BMI<=39.9:
        print(name +",Obesity Stage 2 ") 
    else:
        print(name +",It's over ")
else:
    print("Enter valid input")

print(BMI)
```
